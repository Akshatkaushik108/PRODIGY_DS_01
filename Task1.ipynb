## Top 10 countries with respect to population

```python
total_popu = data1[data1['Indicator Code'] == 'SP.POP.TOTL']
total_sorted = total_popu.sort_values(by = "2023", ascending = False)
total_top10 = total_sorted.head(10)
print("Top 10 countries of total population: \n")
print(total_top10[["Country Name"]])
```

## Plotting

```python
plt.figure(figsize = (15,6))
plt.subplot(2,2,1)
sns.barplot(x = "2023" , y = "Country Name", data = total_top10, palette = "muted")
plt.title("Top 10 countries of total population in 2023", fontsize = 8)
plt.xlabel("Population", fontsize = 7.5)
plt.ylabel("Country", fontsize = 7.5)
plt.show()
```

```python
plt.figure(figsize = (15,6))
plt.subplot(2,2,2)
sns.barplot(x = "1960" , y = "Country Name", data = total_top10, palette = "muted")
plt.title("Top 10 countries of total population in 1960", fontsize = 8)
plt.xlabel("Population", fontsize = 7.5)
plt.ylabel("Country", fontsize = 7.5)
plt.show()
```

## Countries with highest male population

```python
data2=pd.read_csv("malepopulation.csv")
```

```python
data2.head(5)
```

```python
total_male_popu = data2[data2['Indicator Code'] == 'SP.POP.TOTL.MA.IN']
total_male_sorted = total_male_popu.sort_values(by = "2022", ascending = False)
total_male_top10 = total_male_sorted.head(10)
print("Top 10 countries of highest male population: \n")
print(total_male_top10[["Country Name"]])
```

```python
plt.figure(figsize = (15,6))
plt.subplot(2,2,1)
sns.barplot(x = "2022" , y = "Country Name", data = total_male_top10, palette = "cubehelix")
plt.title("Top 10 countries with highest male population in 2022", fontsize = 8)
plt.xlabel("Population", fontsize = 7.5)
plt.ylabel("Country", fontsize = 7.5)
plt.show()
```

```python
data3=pd.read_csv("femalepopulation.csv")
```

```python
data3.head(5)
```

```python
total_fe_popu = data3[data3['Indicator Code'] == 'SP.POP.TOTL.FE.IN']
total_fe_sorted = total_fe_popu.sort_values(by = "2022", ascending = False)
total_fe_top10 = total_fe_sorted.head(10)
print("Top 10 countries with highest male population: \n")
print(total_fe_top10[["Country Name"]])
```

```python
plt.figure(figsize = (15,6))
plt.subplot(2,2,1)
sns.barplot(x = "2022" , y = "Country Name", data = total_fe_top10, palette = "cubehelix")
plt.title("Top 10 countries with highest female population in 2022", fontsize = 8)
plt.xlabel("Population", fontsize = 7.5)
plt.ylabel("Country", fontsize = 7.5)
plt.show()
```

```python
merge_data=pd.merge(total_male_popu,total_fe_popu, on = "Country Code", suffixes = ("_male","_female"))
```

```python
merge_data["Total Population"] = merge_data["2022_male"] + merge_data["2022_female"]
```

```python
merge_data.head(10)
```

```python
sorted_merge_data = merge_data.sort_values(by = "Total Population", ascending = False)
```

```python
top_10_countries = sorted_merge_data.head(10)
```

```python
sorted_merge_data.head(10)
```

```python
plt.figure(figsize = (11,6))
sns.barplot(x="Country Code", y="2022_female", data= top_10_countries, color= "green", label="Female Population")
sns.barplot(x="Country Code", y="2022_male", data= top_10_countries, bottom= top_10_countries["2022_female"], color = "yellow", label = "Male Population")
plt.xlabel("Country")
plt.ylabel("Population")
plt.legend()
plt.show()
```
